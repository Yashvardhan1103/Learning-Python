## Array Operations

```python
# we can do mathematical operations on arrays (addition, subtraction, square, exponents)
# we can also do matrix opeations(product, transpose, inverse, etc.)
```

```python
# arithmetic operators apply element wise

a = np.array([1, 2, 3, 4])
b = np.array([10, 20, 30, 40])

c = a + b
print(c)

d = a * b
print(d)
```

```python
# example - convert a list of farheneit values to celcius
a = np.array([97, 100, 105])
b = (a - 32) * (5 / 9)
print(b)

# now, check whether the celcius values are greater than 37 (if True, then fever)
print(b > 37)

# check if number is even or odd
print(a % 2 == 0)
```

```python
# matrix manipulations - matrix multiplication
m1 = np.array([[1, 1], [0, 1]])
m2 = np.array([[2, 0], [3, 4]])
print(m1 * m2) # this is not a matrix multiplier, * is an arithmetic operator so it only performs element wise
print(m1 @ m2) # this is the dot product of the two matrices
```

```python
# when the arrays are of different types, their manipulation results in corresponding to the more common data type
arr1 = np.array([[1, 2, 3], [4, 5, 6]])
arr2 = np.array([[7.1, 8.2, 9.6], [9.5, 5.3, 2.1]])
arr3 = arr1 + arr2
print(arr3)
print(arr3.dtype)
# we can see that the first array (int) got converted to float
```

```python
# aggregation functions of numpy arrays
print(arr3.sum())
print(arr3.max())
print(arr3.min())
print(arr3.mean())
```

```python
# example - create an array with 15 elements and arrange them in a 3x5 matrix
arr4 = np.arange(1, 16, 1).reshape(3, 5)
print(arr4)
```

## Indexing, Slicing and Iterating

### Indexing

```python
# in a 1D array, works similar to lists
a = np.array([1, 5, 3, 8])
print(a[3])
```

```python
# for a multidimensional array, we use integer array indexing
b = np.array([[1, 2], [3, 4], [5, 6]])
print(b)
print(b[1, 1]) # for element in second row, second column
```

```python
# print multiple values in an existing 2D array into a 1D array
c = np.array([b[0,0], b[0,1], b[2,1]])
print(c)
```

```python
# can also be done using this syntax
print(b[[0, 0, 2], [0, 1, 1]])
```

```python
# boolean indexing
print(b > 3)
```

```python
print(b[b > 3])
# inserts all values that return true for condition inside [] into a 1D array
#this functionality is essential for the Pandas toolkit
```

### Slicing

```python
# For 1D arrays, slicing works similar to lists
arr = np.array([1, 3, 4, 6, 8])
print(arr[:3])
print(arr[2:4])
print(arr[3:])
```

```python
# For multidimensional arrays
arr = np.arange(1, 13, 1).reshape(3, 4)
print(arr)

# for only one argument
print(arr[0:2]) # we get elements from row 0 and 1

#for two arguments
print(arr[:2, 1:3]) # we get elements from row 0 and 1 AND column 1 and 2
```

```python
# if we create a subarray using slicing, and modify a value in it, it will also modify the value in the original array
subarray = arr[:2]
subarray[1, 3] = 50
print(subarray)
print(arr) # value changed
```

### [Trying NumPy with Datasets]

```python
# we are using the wine quality dataset
# to load a dataset in numpy, we can use genfromtxt() function
wines = np.genfromtxt('datasets/winequality.csv', delimiter = ';', skip_header = 1)
print(wines)
```

```python
# now we want to get the fixed acidity of the wines (column 1)
# first parameter in slicing is row, second is column
print(wines[:, 0]) # this would make them appear in a 1D array
print(wines[:, 0:1]) # this would make them appear in their own rows
```

```python
# print the first three columns
print(wines[:, 0:3])

# to view non consecutive columns, we can add the column numbers in a list and pass that list as the second argment
print(wines[:, [2, 4, 5]])
```

```python
# find the average quality of the wines
q = wines[:, -1]
print(q.mean())
```

```python
# now, we will work with the admissionpredict.csv dataset

admission = np.genfromtxt('datasets/admissionpredict.csv', dtype = float, delimiter = ',', skip_header = 1, names = ('Serial No.', 'GRE_Score', 'TOEFL_Score', 'University_Rating','SOP', 'LOR', 'CGPA', 'Research', 'Chance_of_Admit'))
print(admission)
print(admission.shape)
```

```python
# now, we have to convert the GPAs of the studdents to /10 to /4
admission['CGPA'] = (admission['CGPA'] / 10) * 4
print(admission['CGPA'][0:20])
print(admission[0:10])
# the values have been changed in the underlying array as well
```

```python
# now we have to find out how many students have research experience
count = 0
for i in admission['Research'] == 1:
    if i == True:
        count += 1
print(count, '/', len(admission), 'students have research experience')
```

```python
# find if students with admission chance > 0.8 have higher avg gre score than admission chance < 0.4

print(admission[admission['Chance_of_Admit'] > 0.8]['GRE_Score'].mean())
print(admission[admission['Chance_of_Admit'] < 0.4]['GRE_Score'].mean())
```

```python
# also do this for CGPA
print(admission[admission['Chance_of_Admit'] > 0.8]['CGPA'].mean())
print(admission[admission['Chance_of_Admit'] < 0.4]['CGPA'].mean())
```
